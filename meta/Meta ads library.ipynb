```python
import requests

# Get app access token using client credentials
def get_app_access_token(app_id, app_secret):
    url = "https://graph.facebook.com/oauth/access_token"
    params = {
        "client_id": app_id,
        "client_secret": app_secret,
        "grant_type": "client_credentials"
    }
    
    response = requests.get(url, params=params)
    
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error: {response.status_code}")
        print(response.json())
        return None

# Use the app credentials from the first cell
app_access_token = get_app_access_token(app_id, app_secret)
print(app_access_token)
```

```python
# Search Facebook Ads Archive for sustainability ads
def search_ads_archive(access_token, search_terms="sustainability", ad_type="ALL", countries=["US"]):
    url = "https://graph.facebook.com/v19.0/ads_archive"
    
    params = {
        "search_terms": search_terms,
        "ad_type": ad_type,
        "ad_reached_countries": countries,
        "fields": "id,ad_creative_body,ad_snapshot_url,publisher_platform",
        "access_token": access_token
    }
    
    response = requests.get(url, params=params)
    
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error: {response.status_code}")
        print(response.json())
        return None

# Example usage - you'll need to get the app access token first from the previous cell
ads_data = search_ads_archive(app_access_token['access_token'])
```
